# House of prices

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

## Load the dataset

```python
from sklearn.datasets import load_boston
boston=load_boston()
```

```python
#a set-like object providing a view on boston's keys
boston.keys()
```

```python
## Description of the dataset
print(boston.DESCR)
```

```python
#data points inside list
print(boston.data)
```

```python
#sales price inside list
print(boston.target)
```

```python
#all feature names
print(boston.feature_names)
```

## Preparing the Dataset

```python
#columns=boston.feature_names is use to name each columns
dataset=pd.DataFrame(boston.data,columns=boston.feature_names)
```

```python
dataset.head()
```

```python
dataset['Price']=boston.target
```

```python
#first 5 entries
dataset.head()
```

```python
dataset.info()
```

```python
#summarizing stats of data
dataset.describe()
```

```python
## checking missing values
dataset.isnull().sum()
```

```python
## Analysing the Correlated Features using heat map
corr=dataset.corr()
plt.figure(figsize=(11,11))
sns.heatmap(corr,annot=True)
plt.show()
```

```python
for column in dataset.columns:
    if column!="Price":
        plt.scatter(dataset[column],dataset["Price"])
        plt.xlabel(column)
        plt.ylabel("Price")
        plt.show()

```

## Observation

#### 1)RM,LSTAT highly corelated with price
#### 2)INDUS,NOX,TAX,PTRATIO moderatly co-related
#### 3)Rest all Non-corelated/Less corelated
#### 4)When CRIM(crime) is increasing price is decreasing
#### 5)when RM(room) increasing price is increasing



```python
for column in dataset.columns:
    if column!="Price":
        sns.regplot(x=column,y="Price",data=dataset)
        plt.xlabel(column)
        plt.ylabel("Price")
        plt.show()
```

## Observation

#### 1)We have drawn best best-fit line with RM,LSTAT  wrt Price

```python
#Independent and Dependent features
X=dataset.iloc[:,:-1]
y=dataset.iloc[:,-1]
```

```python
X.head()
```

```python
y
```

```python
##Train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.30,random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
#standardize the dataset
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

## Model

```python
from sklearn.linear_model import LinearRegression
regression=LinearRegression()
```

```python
regression.fit(X_train,y_train)
```

```python
## print the coeeicients and the intercept
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
# on which model parameters the model has been trained
regression.get_params()
```

```python
## Predection with test Data
reg_pred=regression.predict(X_test)
reg_pred
```

```python
#plot a scatter for prediction
plt.scatter(y_test,reg_pred)
```

```python
residuals=y_test-reg_pred
residuals
```

```python
#Ploting Residuals
sns.displot(residuals,kind='kde')
```

```python
#scatter plot with respect to prediction and residuals
#uniform distribution
plt.scatter(reg_pred,residuals)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
print(mean_absolute_error(y_test,reg_pred))
print(mean_squared_error(y_test,reg_pred))
print(np.sqrt(mean_squared_error(y_test,reg_pred)))
```

## R-square and adjusted r-square

```python
from sklearn.metrics import r2_score
r2_score=r2_score(y_test,reg_pred)
r2_score
```

```python
#adjusted r-square=1-[(1-r2)*(n-1)/(n-k-1)]
1-(1-r2_score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```

## New Data Prediction

```python
#reshape (13,) to (1,13)
new=scaler.transform(boston.data[0].reshape(1,-1))
```
